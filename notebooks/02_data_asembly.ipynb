Combining the crash and weather datasets

```python
import pandas as pd
```

```python
# 1. Load Data
crash_df = pd.read_csv('../data/processed/cleaned_crashes.csv') 
weather_df = pd.read_csv('../data/processed/cleaned_weather.csv')


# 2. Alignment
crash_df['DATE'] = pd.to_datetime(crash_df['DateOfIncident'])  
crash_df['DATE'] = crash_df['DATE'].dt.normalize() # Normalize to remove time component
weather_df['DATE'] = pd.to_datetime(weather_df['DATE']).dt.normalize() # Normalize weather date column

# 3. Merge Datasets
master_df = pd.merge(
    crash_df,
    weather_df,
    on='DATE',
    how='left'
)

# 4. Output Data
pd.set_option('display.max_columns', None)
master_df = master_df.sort_values(by='DATE', ascending=True)
master_df.sample(5)
```
